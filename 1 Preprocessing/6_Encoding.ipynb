```python
'''
    Encoding Categorical Data -->

    Categorical encoding refers to the process of converting categorical data
    (data that can take on a limited number of values,typically representing categories or labels)
    into a numerical format that machine learning algorithms can understand and work with

    Since many machine learning models can only handle numerical data, categorical features need to be
    transformed into numbers. There are several techniques to achieve this, depending on the type of categorical
    data and the specific problem. Here are some common methods for categorical encoding
'''
```

```python
#   Methods -->

#   Label Encoding
#   One-Hot Encoding
#   Binary Encoding
#   Frequency Encoding
#   Mean Encoding
#   Ordinal Encoding
```

```python
import numpy as np
import pandas as pd
```

```python
'''
    Label Encoding -->

    Label encoding is another technique for converting categorical data into numerical form,
    but unlike one-hot encoding, it assigns an integer to each category. Each unique category
    is given a distinct integer value, which allows algorithms to process the data in numerical form.

    Example -->

    If you have the categories (Red, Blue, Green), label encoding would assign -->

    Red = 0
    Blue = 1
    Green = 2
'''
```

```python
#   Import Label Encoder From sklearn -->

from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('Data/Data.csv')
dataset
```

```python
encoder = LabelEncoder()
dataset['Country'] = encoder.fit_transform(dataset['Country'])
dataset
```

```python
'''
    One Hot Encoding -->

    One-hot encoding is a technique used to represent categorical data as binary vectors.
    It transforms categorical variables, which might not have a numerical relationship,
    into a format that can be provided to machine learning algorithms.

    Here's how it works ->

    Each category in the data is represented by a vector of 0s and 1s.
    The length of the vector equals the number of unique categories.
    For a given category, the corresponding position in the vector is marked as 1, and all other positions are marked as 0.

    Example -->

    Suppose you have a dataset with three categories:
    Red, Blue, and Green. One-hot encoding would represent these categories as:

    Red: [1, 0, 0]
    Blue: [0, 1, 0]
    Green: [0, 0, 1]
'''
```

```python
data = pd.read_csv('Data/Data.csv')
data
```

```python
x_data = data.iloc[:, :-1].values
y_data = data.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
clt = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
x_data = np.array(clt.fit_transform(x_data))
x_data
```

```python
y_data = encoder.fit_transform(y_data)
y_data
```
