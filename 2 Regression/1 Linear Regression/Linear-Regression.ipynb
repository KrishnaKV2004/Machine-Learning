```python
'''
    Simple Linear Regression ->

    Linear regression is a fundamental statistical technique used to model the relationship between
    a dependent variable (also called the target or outcome) and one or more independent variables
    (also called predictors or features). It assumes that the relationship between the variables is linear,
    meaning that changes in the independent variables lead to proportional changes in the dependent variable
'''
```

```python
'''
    Equation -->

    y = β0 + β1x + ϵ

    y = Dependent Variable
    x = Independent Variable
    β0 = Intercept
    β1 = Slope
    ϵ = Error Term
'''
```

```python
#   Importing Libraries -->

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#   Importing Dataset -->

data = pd.read_csv('Data/Salary_Data.csv')
data.head()
```

```python
x_data = data.iloc[:, :-1].values
y_data = data.iloc[:, -1].values
```

```python
#   Splitting Dataset -->

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 0)
```

```python
#   Training Linear Regression Model -->

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#   Predicting Test Set Results -->

y_pred = reg.predict(x_test)
```

```python
#   Visualising Training Set Results -->

plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, reg.predict(x_train), color = 'blue')
plt.title("Salary Prediction")
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#   Visualising Test Set Results -->

plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, reg.predict(x_train), color = 'blue')
plt.title("Salary Prediction")
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.show()
```
