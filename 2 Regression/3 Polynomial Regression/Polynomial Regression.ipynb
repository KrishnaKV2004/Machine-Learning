```python
'''
    Polynomial Regression -->

    Polynomial regression is a form of regression analysis that models the relationship between the independent variable
    x and the dependent variable y. y as an n-th degree polynomial. It is useful when data shows a curvilinear relationship
    rather than a straight line, as linear regression does.
'''
```

```python
'''
    Polynomial Regression -->

    y = β0 + β1x + β2x^2 + β3x^3 + ⋯ + βnx^n + ϵ

    Where -->

    y is the dependent variable (target)
    x is the independent variable (input)
    β0, β1, …, βn are the coefficients to be learned
    n is the degree of the polynomial
    ϵ is the error term
'''
```

```python
#   Importing Libraries -->

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
#   Importing Dataset -->

data = pd.read_csv('Data/Position_Salaries.csv')
data
```

```python
x_data = data.iloc[:, 1: -1].values
y_data = data.iloc[:, -1].values
```

```python
#   Building Linear Model -->

linear_model = LinearRegression()
linear_model.fit(x_data, y_data)
```

```python
#   Applying Polynomial Features To Data -->

polynomial_model = PolynomialFeatures(degree = 4)
x_poly = polynomial_model.fit_transform(x_data)
x_poly
```

```python
#   Building Polynomial Model -->

model = LinearRegression()
model.fit(x_poly, y_data)
```

```python
#   Visualizing Linear Model -->

plt.scatter(x_data, y_data, color = 'red')
plt.plot(x_data, linear_model.predict(x_data), color = 'blue')
plt.title("Linear Model")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
#   Visualizing Polynomial Model -->

plt.scatter(x_data, y_data, color = 'red')
plt.plot(x_data, model.predict(x_poly), color = 'blue')
plt.title("Polynomial Model")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
#   Predicting Salary with Linear Model -->

linear_model.predict([[6.5]])
```

```python
#   Linear Model has very high Error Lets check with Polynomial Model -->

model.predict(polynomial_model.fit_transform([[6.5]]))
```

```python
'''
   Summary -->

   Polynomial regression is linear because it is linear in the coefficients, even though it involves polynomial features
   Degree 4 is chosen when the data shows more complex relationships than can be captured by a lower degree
   (like degree 2 or 3), but care should be taken to avoid overfitting
   Data transformation (e.g., creating polynomial features like x^2, x^3) allowing a linear regression model
   to handle more complex patterns in the data
'''
```
